```python
import pandas as pd

# Adding sample data to reflect the requested analysis areas
data = {
    'Order_date': pd.to_datetime(['2023-01-15', '2023-01-20', '2023-02-10', '2023-02-15', '2023-03-05', '2023-03-20']),
    'Products': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Monitor'],
    'Quantity': [1, 5, 1, 2, 10, 2],
    'Price': [1200, 25, 1200, 75, 25, 300],
    'Profit': [200, 50, 200, 30, 100, 80],
    'Customers': ['Jone', 'Alice', 'John', 'Barry', 'Nana', 'Bob']
}

df = pd.DataFrame(data)
# Calculate Sales column
df['Total price/Sales'] = df['Quantity'] * df['Price']
print(df)
#
```

```python
# 1. Best-selling products
Best_selling = df.groupby('Products')['Quantity'].sum().sort_values(ascending=False)
print("Best-Selling Products:\n", Best_selling)

# 2. Monthly sales trends
monthly_sales = df.set_index('Order_date').resample('M')['Total price/Sales'].sum()
print("\nMonthly Sales Trends:\n", monthly_sales)

# 3. Revenue and Profit analysis
total_revenue = df['Total price/Sales'].sum()
total_profit = df['Profit'].sum()
print(f"\nTotal Revenue: ${total_revenue}")
print(f"Total Profit: ${total_profit}")
```
